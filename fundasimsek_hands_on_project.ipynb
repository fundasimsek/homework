# sklearn kütüphanesi üzerinden LabelEncoder fonksiyonunu import ediniz ve label adındaki değişkene atayınız.Verisetiniz içindeki AIRLINE, ORIGIN_AIRPORT ve DESTINATION_AIRPORT değişkenlerine LabelEncoder transform işlemlerini uygulayınız.

```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
```

```python
flightdata["AIRLINE"] = label.fit_transform(flightdata["AIRLINE"])
flightdata["ORIGIN_AIRPORT"] =label.fit_transform(flightdata["ORIGIN_AIRPORT"])
flightdata["DESTINATION_AIRPORT"] = label.fit_transform(flightdata["DESTINATION_AIRPORT"])
```

# sklearn kütüphanesi üzerinden LogisticRegression ve metrics fonksiyonlarını tanımlayınız.

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

# Guided project üzerinde yapılan işlemlerden yola çıkarak, y ve X değerlerini verisetinizin içindeki GECIKME sütunu için hazırlayınız.

```python
flightdata.isnull().sum()
```

```python
X=flightdata.drop(["GECIKME","AIR_TIME","TAXI_OUT","WHEELS_OFF"],axis=1)
y=flightdata["GECIKME"]
#MISSING VALUE HATASI VERECEĞİ İÇİN NULL DEĞERLER İÇEREN DEĞİŞKENLER ÇIKARILARAK X BELİRLENMİŞTİR.
```

# Verisetinizi modele uygun şekilde train-test olarak ayrabilmek için ilgili fonskiyonları tanımlayınız ve aşağıdaki parametreleri kullanarak train-test verilerini oluşturunuz. (test_size=0.20 , random_state=1933)

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=1933)
```

```python
logReg = LogisticRegression()
```

```python
logReg.fit(X_train,y_train)
```

```python
y_pred = logReg.predict(X_test)
```

# LogisticRegression adımlarını takip ederek "Accuracy of logistic regression classifier on test set:" değerini hesaplayınız.

```python
print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(logReg.score(X_test, y_test)))
```

#Son adım olarak classification_report fonksiyonunu import ederek kurduğunuz modelin sonuçlarını gösteriniz.

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pred))
```
